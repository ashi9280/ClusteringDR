### Relevant Imports

```python
# imports
import sklearn
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from sklearn.mixture import GaussianMixture

from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import SpectralClustering

from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
from sklearn.random_projection import GaussianRandomProjection
from sklearn.decomposition import KernelPCA

from sklearn.datasets import load_breast_cancer #(569, 30, 2)
from sklearn.datasets import load_wine #(178, 13, 3)
from sklearn.datasets import load_iris #(150, 4, 3)
from sklearn.datasets import load_diabetes #(442, 10, 322)
from sklearn.datasets import load_digits #(1797, 64, 10)

from sklearn.datasets import fetch_covtype #(581012, 54, 7)
from sklearn.datasets import fetch_lfw_people #(13233, 5828, 5749)

from sklearn.metrics import accuracy_score
from sklearn.metrics import homogeneity_score

from sklearn.manifold import TSNE
```

### Selecting Datasets

I want to select a high dimensionality and a low dimensionality set. I used cover type (54) and lfw (5828).

```python
# Creating datasets
dataset1 = load_breast_cancer()
dataset2 = load_digits()

data1 = dataset1.data
data2 = dataset2.data

target1 = dataset1.target
target2 = dataset2.target

numTargets1 = len(np.unique(dataset1.target))
numTargets2 = len(np.unique(dataset2.target))
```

```python
# Dummy dataset
testDataset1 = load_iris()
tData1 = testDataset1.data
tTarget1 = testDataset1.target

tnumTargets1 = len(np.unique(testDataset1.target))
```

```python
# Make Predictions
em1Pred = em1.predict(data1)
em2Pred = em2.predict(data2)
em3Pred = em3.predict(tData1)

km1Pred = km1.predict(data1)
km2Pred = km2.predict(data2)
km3Pred = km3.predict(tData1)

"""
dbs1Pred = dbs1.fit_predict(data1)
dbs2Pred = dbs2.fit_predict(data2)
dbs3Pred = dbs3.fit_predict(tData1)

sc1Pred = sc1.fit_predict(data1)
sc2Pred = sc2.fit_predict(data2)
sc3Pred = sc3.fit_predict(tData1)
"""
```

### Dimension Reduction Algorithms

```python
# 
goalDim1 = 6
goalDim2 = 10
goalDim3 = 2
```

```python
# Configure PCA
pca_data1 = PCA(n_components=goalDim1).fit_transform(data1)
pca_data2 = PCA(n_components=goalDim2).fit_transform(data2)
pca_tData1 = PCA(n_components=goalDim3).fit_transform(tData1)
```

```python
# Configure ICA
ica_data1 = FastICA(n_components=goalDim1).fit_transform(data1)
ica_data2 = FastICA(n_components=goalDim2).fit_transform(data2)
ica_tData1 = FastICA(n_components=goalDim3).fit_transform(tData1)
```

```python
# Configure Random Projections
rp_data1 = GaussianRandomProjection(n_components=goalDim1).fit_transform(data1)
rp_data2 = GaussianRandomProjection(n_components=goalDim2).fit_transform(data2)
rp_tData1 = GaussianRandomProjection(n_components=goalDim3).fit_transform(tData1)
```

```python
# Configure Kernel PCA
kpca_data1 = KernelPCA(n_components=goalDim1, kernel='poly', degree=3).fit_transform(data1)
kpca_data2 = KernelPCA(n_components=goalDim2, kernel='poly', degree=3).fit_transform(data2)
kpca_tData1 = KernelPCA(n_components=goalDim3, kernel='poly', degree=3).fit_transform(tData1)
```

```python
# Configure EM on reduced datasets
em1_pca = GaussianMixture(n_components=numTargets1, random_state=1).fit(pca_data1)
em2_pca = GaussianMixture(n_components=numTargets2, random_state=1).fit(pca_data2)
em3_pca = GaussianMixture(n_components=tnumTargets1, random_state=1).fit(pca_tData1)

em1_ica = GaussianMixture(n_components=numTargets1, random_state=1).fit(ica_data1)
em2_ica = GaussianMixture(n_components=numTargets2, random_state=1).fit(ica_data2)
em3_ica = GaussianMixture(n_components=tnumTargets1, random_state=1).fit(ica_tData1)

em1_rp = GaussianMixture(n_components=numTargets1, random_state=1).fit(rp_data1)
em2_rp = GaussianMixture(n_components=numTargets2, random_state=1).fit(rp_data2)
em3_rp = GaussianMixture(n_components=tnumTargets1, random_state=1).fit(rp_tData1)

em1_kpca = GaussianMixture(n_components=numTargets1, random_state=1).fit(kpca_data1)
em2_kpca = GaussianMixture(n_components=numTargets2, random_state=1).fit(kpca_data2)
em3_kpca = GaussianMixture(n_components=tnumTargets1, random_state=1).fit(kpca_tData1)
```

```python
# Configure K means on reduced datasets
km1_pca = KMeans(n_clusters=numTargets1, random_state=1).fit(pca_data1)
km2_pca = KMeans(n_clusters=numTargets2, random_state=1).fit(pca_data2)
km3_pca = KMeans(n_clusters=tnumTargets1, random_state=1).fit(pca_tData1)

km1_ica = KMeans(n_clusters=numTargets1, random_state=1).fit(ica_data1)
km2_ica = KMeans(n_clusters=numTargets2, random_state=1).fit(ica_data2)
km3_ica = KMeans(n_clusters=tnumTargets1, random_state=1).fit(ica_tData1)

km1_rp = KMeans(n_clusters=numTargets1, random_state=1).fit(rp_data1)
km2_rp = KMeans(n_clusters=numTargets2, random_state=1).fit(rp_data2)
km3_rp = KMeans(n_clusters=tnumTargets1, random_state=1).fit(rp_tData1)

km1_kpca = KMeans(n_clusters=numTargets1, random_state=1).fit(kpca_data1)
km2_kpca = KMeans(n_clusters=numTargets2, random_state=1).fit(kpca_data2)
km3_kpca = KMeans(n_clusters=tnumTargets1, random_state=1).fit(kpca_tData1)
```

```python
# Configure SC on reduced datasets
"""
sc1_pca = SpectralClustering(n_clusters=numTargets1)
sc2_pca = SpectralClustering(n_clusters=numTargets2)
sc3_pca = SpectralClustering(n_clusters=tnumTargets1)

sc1_ica = SpectralClustering(n_clusters=numTargets1)
sc2_ica = SpectralClustering(n_clusters=numTargets2)
sc3_ica = SpectralClustering(n_clusters=tnumTargets1)

sc1_rp = SpectralClustering(n_clusters=numTargets1)
sc2_rp = SpectralClustering(n_clusters=numTargets2)
sc3_rp = SpectralClustering(n_clusters=tnumTargets1)

sc1_kpca = SpectralClustering(n_clusters=numTargets1)
sc2_kpca = SpectralClustering(n_clusters=numTargets2)
sc3_kpca = SpectralClustering(n_clusters=tnumTargets1)

sc1_pca.fit(pca_data1)
sc2_pca.fit(pca_data2)
sc3_pca.fit(pca_tData1)

sc1_ica.fit(ica_data1)
sc2_ica.fit(ica_data2)
sc3_ica.fit(ica_tData1)

sc1_rp.fit(rp_data1)
sc2_rp.fit(rp_data2)
sc3_rp.fit(rp_tData1)

sc1_kpca.fit(kpca_data1)
sc2_kpca.fit(kpca_data2)
sc3_kpca.fit(kpca_tData1)
"""
```
